```python
# Criar uma escala de valores em milhões

dados_vendas['Valor em Milhoes'] = dados_vendas['Valor_anuncio_float'].apply(lambda a : a/1000000)
dados_vendas[['Valor_anuncio_float', 'Valor em Milhoes']]
```

```python
# Deixar o gráfico do histograma de valores legível (alterar labels, cores, título, escala)

sns.set()
plt.figure(figsize=(10, 8))
ax = sns.histplot(data = dados_vendas, x = "Valor_anuncio_float", kde = True)
ax.set_title("Histograma Valor Imóvel")
plt.xlabel("Valor do Anúncio")
plt.ylabel("Contagem")
plt.xlim((-50, 10000000))
plt.show()
```

```python
# Preço do metro quadrado por bairro e plotar em um gráfico ideal

m2 = dados_vendas[['Bairro', 'Metragem', 'Valor_anuncio_float']]
m2
```

```python
m2 = dados_vendas[['Bairro', 'Metragem', 'Valor_anuncio_float']].groupby('Bairro', as_index=False).sum()
m2
```

```python
m2['valor_m2'] = m2['Valor_anuncio_float']/m2['Metragem']
m2
```

```python
m2.sort_values(by='valor_m2', ascending=False, inplace=True)
```

```python
m2[0:20]
```

```python
sns.set()
plt.figure(figsize=(10, 8))
ax = sns.barplot(data = m2[:20], x = 'valor_m2', y='Bairro', orient='h')
ax.set_title('Os 20 metros quadrados mais caros em São Paulo')
plt.xlabel('Valor do metro quadrado')
plt.show()
```
